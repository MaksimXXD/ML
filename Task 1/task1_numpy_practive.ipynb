 
 
* __Задача 6__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Дополнительно сравните с функцией scipy.spatial.distance.cdist по скорости работы (сравнения приведите ниже в ноутбуке).

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time
import functions
import functions_vectorized
import scipy
%matplotlib inline
ans = []
ans_np = []
ans_sdc = []
data_size = []
data_size1= []
k = 6
for i in range(10,7000,100):
    n = i
    data_size.append(n)
    arr1 = np.random.randint(0,100,size=(n,k))
    arr2 = np.random.randint(0,100,size=(n,k))
    if (i <= 800):
        data_size1.append(n)
        start = time.time()
        functions.pairwise_distance(arr1,arr2)
        end1 = time.time() - start
        ans.append(end1)
    start = time.time()
    functions_vectorized.pairwise_distance(arr1,arr2)
    end2 = time.time() - start
    start = time.time()
    scipy.spatial.distance.cdist(arr1, arr2)
    end3 = time.time() - start
    ans_np.append(end2)
    ans_sdc.append(end3)
plt.plot(data_size1, ans, label=u"Невекторизованная")
plt.plot(data_size, ans_np, label=u"Векторизованная")
plt.plot(data_size, ans_sdc, label=u"scipy")
plt.title(u"Сравнение времени выполнения вектор и невектор")
plt.xlabel(u"Размер данных")
plt.ylabel(u"Время (с)")
plt.grid()
plt.legend()
```
